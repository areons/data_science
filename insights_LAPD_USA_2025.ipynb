```python
import pandas
import numpy
df=pandas.read_csv('Crime_Data_from_2020_to_Present.csv')
df
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df=df.fillna(0)
```

```python
df.drop(['LOCATION', 'Cross Street', 'LON', 'LAT'], axis=1, inplace=True)
```

```python
df=df.fillna(0)
df.isnull().sum()
```

```python
df.drop(['DR_NO', 'AREA', 'Mocodes', 'Crm Cd 3'], axis=1, inplace=True)
```

```python
df.drop('Crm Cd 4', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Date Rptd'] = pandas.to_datetime(df['Date Rptd'])
df['DATE OCC'] = pandas.to_datetime(df['DATE OCC'])
```

```python
df.groupby(['Crm Cd Desc']).size().reset_index(name='Crm Occurrences').sort_values(by='Crm Occurrences', ascending=False)
```

```python
total = len(df['Status Desc'])
invest_cont = (df['Status Desc'] == 'Invest Cont').sum()
adult_arrest = (df['Status Desc'] == 'Adult Arrest').sum()
pcent_invest_cont = (invest_cont / total) * 100
pcent_adult_arrest = (adult_arrest / total) * 100
print(f'Percentual of Investigations undergoing: {pcent_invest_cont:.0f}%.')
print(f'Percentual of Finished cases: {pcent_adult_arrest:.0f}%.')
```

```python
df1=df.query('`Vict Sex` == "F" or `Vict Sex` == "M"')
```

```python
df2=df1.groupby(['Vict Age', 'Vict Sex']).size().unstack(fill_value=0)
df2['Total'] = df2.sum(axis=1)
```

```python
for col in ['F', 'M']:
    if col not in df2.columns:
        df2[col] = 0
df2['% Difference'] = ((df2['F'] - df2['M']) / df2['M'].replace(0, numpy.nan)) * 100
df2['% Difference'] = df2['% Difference'].fillna(0).round(1)
```

```python
df2[df2.index >= 18].sort_values(by='Total', ascending=False)
```

```python
F_M_avg=df2['% Difference'].mean().round(2)

df_25_35 = df[(df['Vict Age'] >= 25) & (df['Vict Age'] <= 35)]
percent_25_35 = (len(df_25_35) / len(df['Vict Age'])) * 100
print(f"Percentual of preference of victim {F_M_avg}%")
print(f"Total crimes of victims between 25-35 years: {len(df['Vict Age'])} victims")
print(f"Percentual over total: {percent_25_35:.1f}%")
```

```python
df_invest_cont=df[df['Status Desc'] == 'Invest Cont']
df3=df_invest_cont.groupby('Crm Cd Desc').size().reset_index(name='OCCURRENCES').sort_values(by='OCCURRENCES', ascending=False)
df3
```

```python
df.groupby('AREA NAME').size().reset_index(name='occurrences').sort_values(by='occurrences', ascending=False)
```

```python
top_5=df.groupby('AREA NAME').size().reset_index(name='occurrences').sort_values(by='occurrences', ascending=False).head(5)
total_top5 = top_5['occurrences'].sum()
total_geral = df.groupby('AREA NAME').size().sum()
perc_top5 = (total_top5 / total_geral) * 100
print(f"The top 5 zones with most crime occurrences have {perc_top5.round()}% out ot the total occurrences.")
```

```python
df['YEAR']=df['DATE OCC'].dt.year
df['MONTH']=df['DATE OCC'].dt.month
df['DAY']=df['DATE OCC'].dt.day
df.head()
```
