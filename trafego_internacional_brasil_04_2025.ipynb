```python
df1=df.groupby(['UF_ATENDIMENTO', 'TIPO'])['TOTAL'].sum().sort_values(ascending=False)
df1
```

```python
df_group=df1.reset_index()
# criando coluna de porcentagem
df_group['PERCENTUAL'] = (df_group['TOTAL'] / df_group['TOTAL'].sum()) * 100
df_group['PERCENTUAL'] = df_group['PERCENTUAL'].round(2)
df_group
```

```python
df2=df.groupby(['CLASSIFICACAO'])['TOTAL'].sum().sort_values(ascending=False)
df2
```

```python
df_s_outros = df[df['CLASSIFICACAO']!='OUTROS']
df_data=df_s_outros.groupby(['CLASSIFICACAO', 'UF_ATENDIMENTO'])['TOTAL'].sum().reset_index()
df_data
```

```python
df_data['%'] = (df_data['TOTAL'] / df_data['TOTAL'].sum()) * 100
df_data['%'] = df_data['%'].round(2)
df_data.sort_values(by='TOTAL', ascending=False)
```

```python
df_nacionalidade=df.groupby(['NACIONALIDADE', 'TIPO'])['TOTAL'].sum()
df_nacionalidade.sort_values(ascending=False)
```

```python
df_pivot=df_nacionalidade.unstack(fill_value=0)
df_pivot['NET_FLOW'] = df_pivot['ENTRADA'] - df_pivot['SAIDA']
df_pivot.sort_values(by='NET_FLOW', ascending=False)
```
